```python
import OcrToTableTool as ottt
from TableExtractor import TableExtractor as te
import TableLinesRemover as tlr
import cv2
import re
from pdf2image import convert_from_path
import pandas as pd
from google.colab.patches import cv2_imshow
```

```python
# extarct each table
path_to_image = "/content/drive/MyDrive/ocr-extract-table-from-image-python/page0.jpg"
table_extractor = te(image_path=path_to_image,tab_num=3)  # tab_num: number of table in the image. in this image there are three tables
perspective_corrected_image = table_extractor.execute()
for i in range(len(perspective_corrected_image)):
  cv2_imshow(perspective_corrected_image[i])
  print("-"*50)

# OCR on each table
for i in range(len(perspective_corrected_image)):
  lines_remover = tlr.TableLinesRemover(perspective_corrected_image[i])
  image_without_lines = lines_remover.execute()
  ocr_tool = ottt.OcrToTableTool(image_without_lines, perspective_corrected_image[i],f"table_{i}")
  cv2_imshow(image_without_lines)
  ocr_tool.execute()
  print("-"*50)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# read content of each table which is written on  (.csv) format
file1 = pd.read_csv("/content/drive/MyDrive/ocr-extract-table-from-image-python/table_0_output.csv", names=['Benefit', 'Number of Claims', 'Amount of Paid Claims', 'Amount of Claims VAT', 'Notes'])
file2 = pd.read_csv("/content/drive/MyDrive/ocr-extract-table-from-image-python/table_1_output.csv", skiprows=1, names=['Monthly Claims', 'Number of Lives Insured', 'Number of Claims', 'Amount of Paid Claims', 'Amount of Claims VAT'])
file3 = pd.read_csv("/content/drive/MyDrive/ocr-extract-table-from-image-python/table_2_output.csv",  names=['Key', 'Value', 'OtherKey', 'OtherValue'])
```

```python

# File 1 (OverallBenefits): Needs parsing by line due to lack of column headers on each line
file1['Policy Year'] = file1["Benefit"].iloc[1]
file1.drop([0, 1],inplace=True)
file1.reset_index(drop=True, inplace=True)


# File 2 (Monthly Claims): Handle different policy years and totals separately
file2['Policy Year'] = None  # We'll set the Policy Year based later on

# Identify the range of rows for each policy year
prior_policy_rows = file2[file2['Monthly Claims'].str.contains("Prior Policy Year", na=False)].index[0]
policy_year_2_prior_rows = file2[file2['Monthly Claims'].str.contains("Policy Year-2 years prior", na=False)].index[0]
last_policy_year_rows = file2[file2['Monthly Claims'].str.contains("Last Policy Year", na=False)].index[0]
rangeDict = {prior_policy_rows:"Prior Policy Year", policy_year_2_prior_rows:"Policy Year-2 years prior", last_policy_year_rows:"Last Policy Year"}
rangeList = sorted(rangeDict.keys())


# Set Policy Year values for each group
file2.loc[rangeList[0]:rangeList[1], 'Policy Year'] = rangeDict[rangeList[0]]
file2.loc[rangeList[1]:rangeList[2], 'Policy Year'] = rangeDict[rangeList[1]]
file2.loc[rangeList[2]:, 'Policy Year'] = rangeDict[rangeList[2]]

# Remove rows with policy year descriptions
file2 = file2[~file2['Monthly Claims'].str.contains("Prior Policy Year|Policy Year-2 years prior|Last Policy Year", na=False)]

# File 3 (Metadata): Use metadata to extract limits, class, and expiry date
end_date = file3[file3['Key'].str.contains("Policy Expiry Date", na=False)]['Value'].values[0]
class_info = file3[file3['Key'].str.contains("Class", na=False)]['Value'].values[0]
overall_limit = file3[file3['OtherKey'].str.contains("OverallBenefit Limit", na=False)]['OtherValue'].values[0]

end_date_dt = pd.to_datetime(end_date, format='%b %d%Y')
end_date_formatted = end_date_dt.strftime('%Y-%m-%d')


file1['End Date'] = end_date_formatted
file1['Class'] = class_info
file1['Overall Limit'] = overall_limit
file1["Notes"] = file1["Notes"].apply(lambda x: "No info"if str(x)=="nan" else ((re.search(r'(\d+)\s*%', str(x))).group(1) if re.search(r'(\d+)\s*%', str(x)) else ("Yes" if str(x) == "Cesarean is covered" else "No") ))

file2['End Date'] = end_date_formatted
file2['Class'] = class_info
file2['Overall Limit'] = overall_limit


claims_data = file2
benefits_data = file1
```
